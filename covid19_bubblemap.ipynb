```
"""
Created on Tue Mar 17 11:34:02 2020

@author: thanos
"""

import plotly.graph_objects as go
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
df.head()

dates = list(df)
dates = dates[4:np.size(dates)]
```

```
fig = go.Figure()

for i in range(0,len(df)):
    fig.add_trace(go.Scattergeo(
            lon = np.array(df.iloc[i]['Long']),
            lat = np.array(df.iloc[i]['Lat']),
            text = np.array(df.iloc[i][dates[np.size(dates)-1]]),
            name = df.iloc[i]['Country/Region'],
            marker = dict(
                #size = np.array(df.iloc[i][dates[np.size(dates)-1]])/50,
                line_width = 0
            )))
    
fig.show()
```
